$$ Z = \frac{ \biggr[1 + \biggr(\frac{\omega}{\omega_c} \frac{1}{1+\frac{R_1}{R_2}} \biggr)^2 \biggr]R_1 - j\biggr[\frac{\omega}{\omega_c} \frac{\frac{R_1}{R_2}}{1+\frac{R_1}{R_2}}  \biggr]R_1}{1+\biggr( \frac{\omega}{\omega_c}\biggr)^2} $$

$$ |Z| = \sqrt{\frac{\biggr(\biggr[1 + \biggr(\frac{\omega}{\omega_c} \frac{1}{1+\frac{R_1}{R_2}} \biggr)^2 \biggr]R_1\biggr)^2 + \biggr( \biggr[\frac{\omega}{\omega_c} \frac{\frac{R_1}{R_2}}{1+\frac{R_1}{R_2}}  \biggr]R_1\biggr)^2}{\biggr(1+\biggr( \frac{\omega}{\omega_c}\biggr)^2\biggr)^2}} $$

$$ \omega_c \equiv \frac{1}{(R_1 + R_2)C_2}$$


```python
import numpy as np 
import math 
import matplotlib.pyplot as plt

n=10
#t = 0.00005

freq = np.array([100, 280, 460, 640, 820, 1000, 2800, 4600, 6400, 8200, 10000])
omega = 2*np.pi*freq

# collected voltages in mV
V = np.array([13.6, 13.4, 13.5, 13.3, 13.2, 13.0, 10.1, 7.7, 6.1, 5.0, 4.2])

R1 = 200 # Ohms +/- 5%
R2 = 20 # Ohms +/- 5%
C2 = 0.22e-6 # F +/- 10%

omega_c = 1/((R1 + R2)*C2)


Z = np.sqrt(( (R1 + ((omega*R2*R1**2)/(omega_c*R2+omega_c*R1))**2)**2 +  ((omega*R1**2)/(omega_c*R2 + omega_c*R1))**2 ) / ((1 + (omega/omega_c)**2)**2))

phi = np.arctan(-omega_c/omega)

phi_D = phi*180/np.pi 

# divide by V/1000 to convert from mV to V
i = V/(1000*Z)

print(f'Voltages: {V}')
print()
print(f'Impedances given different frequencies: {np.round(Z, -1)}')
print()
print(f'Currents: {np.round(i, 7)}')
print()
print(f'Phase angles (degrees): {np.round(phi_D, 2)}')
```

```python
# preliminary plot for freq vs voltage

plt.plot(freq, V)
plt.show()

# freq vs impedance
plt.plot(freq, Z)
plt.show()
```
