# Impedance Spectroscopy - Ideas

## Note about prior (or just-in-time) knowledge
This experiment depends on familiarity with complex numbers, complex algebra, and the meaning of the complex-valued exponential function. See the appendix for a review or just-in-time learning.

## 1 Purpose of the Experiment

We are going to examine a variety of "systems under test" in which a current $i(t)$ flows through the system and we measure the resulting potential difference $v(t)$ across the system. In particular, we will examine important cases where there is a linear response of the potential to an imposed (co-)sinusoidally varying current where the magnitude and phase of the response can be measured as a function of frequency.

Initially, the system-under-test will literally be a black box containing some simple combination of a resistor, capacitor, and/or inductor. Then we will explore a particular combination of these components that we claim represents the behavior of living tissue. Eventually, we plan to take a step closer to real living materials by using a biological sample (potato slice) whose internal structure gives it an interesting impedance response. This takes us close to the possibility of using such techniques for examining human tissue. The methods you are exploring here are in fact now used in medical diagnostics, including the detection of cancer. See the section: "Point of Departure."

In order to measure the relationship between potential difference and current, both in terms of amplitude and phase, we will introduce a new type of instrument: the lock-in amplifier. This device has both digital and analog displays that directly give these two pieces of information: amplitude and phase. However, the function of the lock-in amplifier goes far beyond this: it is capable of filtering the signal it measures in such a way that it eliminates most of any interfering noise, and thus is a very useful instrument when the currents and potentials are extremely weak and would otherwise be obscured by noise in the system. Initially, we won't need to use weak signals, but in the later part of the experiment where we simulate the measurement of current-voltage characteristics of living tissue, there will be a reduction in signal strength in order to represent physiological conditions - conditions that can also give rise to noisy interference.

The end goal is to gain deep insight into the response of materials to electric fields, where the response includes current flow, electric polarization, and consequent effects on potential drop across a material. These effects determine the electric impedance of the material (defined below), and especially in how the impedance depends on frequency. The details of such dependence contains information not only about the constituent matter but also about the internal structure of a complex medium (tissue, battery, other). Thus the method called "impedance spectroscopy", like its optical counterpart (visible light spectroscopy), uses measurement of the frequency dependence of the response to a probe (injected current) to obtain fundamental insight into a physical system. 

## 2 The idea of complex impedance

![General object whose linear reponse of voltagte to current is to be measured](./DrawingsImages/impedanceObject.png)

As noted above, the experiment will explore the relationship between the electric current $i(t)$ flowing through a system under test and the scalar potential difference $v(t)$ across that system. We will consider the behavior when a (co-)sinusoidal current is imposed with angular frequency $\omega$ radians/second. We'll choose the origin of time $t = 0$ so that at that time the current has zero phase shift in the cosine function:

$$ i(t) = I\cos\omega t. $$

If the system is linear (more on this below), we expect that the corresponding potential difference will also have a (co-)sinusoidal dependence at the same frequency but with a possibly different phase and an amplitude that must be determined as well.

$$ v(t) = V\cos(\omega t + \phi). $$

What do we mean by a linear system? In general we will find that current and voltage are related by linear expressions of a form like

$$ a_m\frac{d^m}{dt^m}v+...+a_1\frac{d}{dt}v+a_0v = b_n\frac{d^n}{dt^n}i+...+b_1\frac{d}{dt}i+b_0i.$$

If $v_1(t)$ is the potential solving this system with current $i_1(t)$ and $v_2(t)$ is the potential solving this system with current $i_2(t)$ then $\alpha v_1(t) + \beta v_2(t)$ is the potential with current $\alpha i_1(t) + \beta i_2(t)$.

Moreover, we can have a complex-valued solution 

$$\tilde v(t) \equiv v_r(t) + j v_i(t)$$

for complex-valued current

$$\tilde i(t) \equiv i_r(t) + j i_i(t)$$

where the subscript $r$ denotes real part, the subscript $i$ denotes imaginary part, and $j \equiv \sqrt{-1}$. 

A key point is that because of the linearity of the system, the real and imaginary parts of the above expressions must separately be solutions: v_r(t) corresponds to i_r(t) and v_i(t) corresponds to i_i(t).

These ideas are important because we can use elegant methods to solve the system in complex form and then take only the real parts at the end to represent the obseved potential and current, since observed quantities are real numbers.

In particular, we can represent the complex solutions usiing complex exponential form, again assuming a (co-)sinusoidal dependence:

$$ \tilde i(t) = Ie^{j\omega t}$$

and


$$ \tilde v(t) = Ve^{j(\omega t +\phi)}.$$

When taking the real parts, we recover our presumed solution for the behavior of the linear physical system:

$$i(t) \equiv i_r(t) =  I\cos\omega t$$

amd 

$$v(t) \equiv v_r(t) = V \cos(\omega t + \phi).$$


Now insert the complex forms of the solutions into the differential equation 3. Notice that the behavior of complex exponentials under differentiation gives results like

$$a_m\frac{d^m}{dt^m}Ve^{j(\omega t +\phi)} = a_mj^m\omega^mVe^{j(\omega t +\phi)}.$$

If we apply this property throughout the differrential equation and cancel the common factor of $e^{j\omega t}$, we can elegantly replace the differential equation with an algebraic one:

$$(a_mj^m\omega^m +...+a_1j\omega + a_0)Ve^{j\phi}=(b_nj^n\omega^n +...+b_1j\omega + b_0)I.$$

(Note the powers of j in the above expression that will simplify to four choices: $j,-j,1,$ and $-1$. For example, $j^4 = (j^2)^2 = (-1)^2 = 1$.

Here we introduce **phasor notation**:

$$\hat V \equiv Ve^{j\phi}$$

and 

$$\hat I \equiv I$$

(More generally, if the current had a nonzero phase $\phi_I$, it would be included so that we have $\hat I \equiv  Ie^{j\phi_I}$.)

Now we can define the **complex impedance** Z of the system:

$$Z \equiv \frac{\hat V}{\hat I}.$$

(We customarily don't write a "hat" above Z: the fact that it might be complex is assumed at the outset.) We see that the magnitude and phase of Z are respectively

$$|Z| = \frac{V}{I}$$

and

$$\text{arg}(Z) = \phi$$.

For our generial lsystem, we see that 

$$Z(\omega) = \frac{b_nj^n\omega^n +...+b_1j\omega + b_0}{a_mj^m\omega^m +...+a_1j\omega + a_0}.$$

We have explicitly shown that we get a complex-valued function of $\omega$. 

$$Z(\omega) = |Z(\omega)|e^{j\phi(\omega)}.$$

The magnitude and phase of the impedance function are each respectively complex functions of $\omega$. We will see how these are determined by looking at explicit examples below.

## 3 Impedance of Basic Electronic Devices

Let us explore the use of the concept of impedance by finding the potential drop $v(t)$ across three fundamental devices assuming a current $i(t) = I \cos\omega t$ is injected into each device. Using complex representations and impedances may seem to be "much ado about nothing", since it is possible to find the potential $v(t)$ across simple resistors, capacitors, or inductors by direct solution of simple algebraic or differential equations. However, when we start combining these components into series and parallel combinations, we will find that the complex impedance approach vastly simplifies the calculations.

### 3.1 Resistor

Consider an ideal resistor. The relation between current $i(t)$ and potential $v(t)$ is:

$$ v = i R. $$

For a current $i = I \cos\omega t$ we see that

$$ v = (I\cos\omega t)R = IR \cos\omega t$$.

Comparing this to our general response

$$v = V cos(\omega t + \phi)$$

we easily see that $V = IR$ and $\phi = 0$, so $\hat V = V = IR = \hat I R$ and conclude that for this particular case 

$$ Z \equiv \frac{\hat V}{\hat I} = R$$.

A more direct way to show this is to find complex solutions for the initial relation:

$$\hat Ve^{j\omega t} = \hat I e^{j \omega t} R$$.

Canceling the common factor of time dependent exponentials,

$$\hat V = \hat I R. $$

So

$$Z \equiv \frac{\hat V}{\hat I} = R,$$

as above.

### 3.2 Capacitor

For a capacitor with capacitance C, we have the fundamental relation between the charge $q$ on the capacitor and the potential difference $v$:

$$ q(t) = C v(t). $$

Take the derivative of both sides and use the fact that current $i = \frac{dq}{dt}$:

$$ i = \frac{dq}{dt} = C\frac{dv}{dt}$$

so our fundamental relation between current and potential in the time domain is

$$ i =  C\frac{dv}{dt}.$$


#### 3.2.1 Finding capacitor impedance the hard way

If $v = V\cos(\omega t + \phi)$ then

$$ i = -\omega C V \sin(\omega t + \phi).$$

$$ i = \omega C V \cos(\omega t + \phi + \frac{\pi}{2}).$$

Since we choose for the current to have zero net phase, we must have

$$ \phi = -\frac{\pi}{2}. $$

We also see the relation between magnitudes

$$ I = \omega CV$$

or 

$$V = \frac{1}{\omega C}I.$$

Thus for

$$i = I \cos \omega t$$,

we find

$$v = \frac{1}{\omega C} I \cos (\omega t - \frac{\pi}{2}) = \text{Real-part-of}\left[\frac{1}{\omega C} I e^{j (\omega t - \frac{\pi}{2})}\right].$$

$$\hat V = \frac{1}{\omega C}e^{-j \frac{\pi}{2}}\space\hat I = -j\frac{1}{\omega C}\hat I = \frac{1}{j\omega C}\hat I.$$

$$Z \equiv \frac{\hat V}{\hat I} = \frac{1}{j\omega C}.$$

#### 3.2.2 FInding capacitor impedance the easy way

Substitute $\hat I e^{j\omega t}$ and $\hat V e^{j\omega t}$ into the equation

$$ i = C\frac{dv}{dt}$$

to get, after cancelling the common factor of $e^{j\omega t}$,

$$ \hat I = C j \omega \hat V$$

and thus

$$Z \equiv \frac{\hat V}{\hat I} = \frac{1}{j\omega C}.$$

Knowing this expression for thecomplex impedance of a capacitor should become second-nature as you proceed in the study of physics.


### 3.3 Inductor

Now we'll quickly cut to the chase:

For an inductor with inductance L,

$$ v = L \frac{di}{dt}.$$

Substitute $\hat V e^{j\omega t}$ and $\hat I e^{j\omega t}$ into the equation and cancel  the common factor of $e^{j\omega t}$ to get:

$$\hat V = L j\omega \hat I.$$

Thus

$$Z \equiv \frac{\hat V}{\hat I} = j\omega L.$$

Again, this is a relation that should become easily-recalled knowledge for a physicist.

## 4 Series and parallel impedances

Recall that for series and parallel combinations of resistors, the equivalent resistances are

$$R_{\text{series }} = R_1 + R_2$$

and 

$$R_{\text{parallel }} = \left(\frac{1}{R_1} + \frac{1}{R_2}\right)^{-1}=\frac{R_1R_2}{R_1+R_2}.$$

These were derived from reasoning
- the same current flows through two components connected in series and the potential differences add
- the same potential difference exists across components connected in parallel and the currents add.

![series and parallel impedances](./drawingsImages/impedanceSeriesParallel.png)

We can apply the same reasoning to complex representations of current and potential to arrive at the comparable rule for series and parallel impedances:

$$Z_{\text{series }} = Z_1 + Z_2$$

$$Z_{\text{parallel }} = \left(\frac{1}{Z_1} + \frac{1}{Z_2}\right)^{-1}=\frac{Z_1Z_2}{Z_1+Z_2}.$$

More complex circuit combinations can be decomposed into a heirarchy of these two cases so that a complete expression can be obtained. We'll see an example in our ultimate circuit-under-test for this experiment.

## Appendix - Review of Complex Numbers

### Pure imaginary number

We generalize our notion of numbers by defining an imaginary constant

$$j \equiv \sqrt{(-1)}.$$

(We use the symbol $j$ instead of $i$ since we will be using $i$ to represent real-valued time-varying electric current.)

Then we see that
$$j^2 = -1.$$

For a real number $y$, we can form an *imaginary number* by multiplying by $j$:

$$jy.$$

### Complex numbers

A complex number $z$ is constructed from two real numbers $x$ and $y$ as:

$$z = x + j y.$$

We say that $x$ is the *real part* of $z$ and that $y$ is the imaginary part of $z$.

### Complex conjugate

The *complex conjugate* of a complex number $z = x + jy$ is denoted with an asterisk $z^*$ and is the number whose imaginary part has opposite sign

$$z^* = x - jy.$$

### Complex addition

Two complex numbers are added in such a way that the real and imaginary parts group separately together: if $z_1 = x_1+jy_1$ and $z_2 = x_2+jy_2$, then

$$z_1 + z_2 = (x_1+jy_1) + (x_2+jy_2) = (x_1 + x_2) + j(y_1 + y_2).$$

The real part of the sum is the sum of the real parts and the imaginary part of the sum is the sum of the imaginary parts.

### Complex multiplication

Complex multiplication requires distributing the individual products and using the fact that 

$$ z_1 \times z_2 = (x_1+jy_1)(x_2+jy_2) = x_1x_2 + j(x_1y_2 + x_2y_1) + j^2y_1y_2 = (x_1x_2 - y_1y_2) +  j(x_1y_2 + x_2y_1).$$

### Complex division

Obtaining the real and imaginary parts of the quotient of two complex numbers requires a step of multiplying both numerator an denominator by the complex conjugate of the denominator:

$$ \frac{z_1}{z_2} = \frac{(x_1+jy_1)}{(x_2+jy_2)}=\frac{(x_1+jy_1)}{(x_2+jy_2)}\frac{(x_2-jy_2)}{(x_2-jy_2)}=\frac{x_1x_2+y_1y_2}{x_2^2+y_2^2}+j\frac{x_2y_1-x_1y_2}{x_2^2+y_2^2}.$$


### Magnitude of a complex number

We compute the magnitude of a complex number z is denoted by |z| and is a real number obtained by taking the square-root of the number multiplied by its complex conjugate:

$$|z| \equiv \sqrt{z^*z} = \sqrt{(x-jy)(x+jy)} = \sqrt{x^2+y^2}.$$

### Phase of a complex number

The phase $\phi$ of a complex number $z = x + jy$ is obtained by:

$$\phi = \tan^{-1}\frac{y}{x}.$$

### Geometric representation of complex numbers

![Complex plane](./drawingsImages/complexPlane.png)

A very important idea is the representation of a complex number by a point in a plane. The coordinate on the horizontal axis is $x$ and the coordinate long the vertical axis is $y$. The magnitude $|z|$ is the magnitude of the vector pointing from the origin to the point with coordinates $(x,y)$ and the phase $\phi$ is the angle the vector makes with the horizontal axis, where a positive value is taken for counterclockwise rotation.

### Euler's formula

An important concept about complex numbers is an identify for the exponential function of an imaginary number called Euler's formula:

$$e^{j\phi} = \cos \phi + j \sin \phi.$$

This can be understood by a Taylor-series exansion

$$ e^{j\phi} = 1 + j\phi + \frac{1}{2}j^2\phi^2 + \frac{1}{6}j^3\phi^3 + ...$$

$$ e^{j\phi} = 1 + j\phi - \frac{1}{2}\phi^2 - \frac{1}{6}j \phi^3 +...$$

$$ e^{j\phi} = (1 - \frac{1}{2}\phi^2 + ...) + j(\phi - \frac{1}{6}\phi^3 +)...$$

$$ e^{j\phi} = \cos \phi + j \sin \phi.$$

### Complex exponential representation of a complex number

The geometric view of a complex number in the x-y plane combined with Euler's formula lets us represent any complex number in the form

$$z = |z|e^{j\phi}.$$

We see this from:

$$ |z|e^{j\phi} = |z|\cos\phi+j|z|\sin\phi = x + jy.$$
