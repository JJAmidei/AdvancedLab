These are a group of sample scatter plots with Gaussian fits with the data we collected on 09/15/2023. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def gauss_function(x, amplitude, mean, stddev):
    return amplitude * np.exp(-(x - mean) ** 2 / (2 * stddev ** 2))

filenames = [
    'Ba-133data.csv',
    'Cd-109data.csv',
    'Co-57data.csv',
    'Co-60data.csv',
    'Mn-54data.csv',
    'Na-22data.csv',
    'Unknowndata.csv'
]

fig, axes = plt.subplots(7, 1, figsize=(8, 18))

for i, filename in enumerate(filenames):
    data = np.loadtxt(filename, skiprows=22, delimiter=',', usecols=(0, 2))
    x = data[:, 0]
    y = data[:, 1]
    
    initial_guess = [1.0, np.mean(x), np.std(x)]
    
    if 'Co-60data.csv' in filename: #
        params, params_covariance = curve_fit(gauss_function, x, y)
    else:
        params, params_covariance = curve_fit(gauss_function, x, y, p0=initial_guess)
    
    amplitude, mean, stddev = params
    
    x_fit = np.linspace(min(x), max(x), 100)
    y_fit = gauss_function(x_fit, amplitude, mean, stddev)
    
    axes[i].scatter(x, y, label='Data Points', color='red', marker='.')
    axes[i].plot(x_fit, y_fit, label='Gaussian Fit', color='blue')
    axes[i].set_xlabel('Channel')
    axes[i].set_ylabel('Count')
    axes[i].set_title(f'{filename}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
