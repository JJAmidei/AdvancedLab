```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import datetime

# Define the Gaussian function
def fgaussian(x, A, B, C, D):
    return A * np.exp(-((x - B) / C)**2) + D

# Create a function to fit multiple Gaussians
def fit_multiple_gaussians(x_data, y_data, initial_guesses):
    # Initialize lists to store parameters and fits
    params_list = []
    fits_list = []

    # Fit each Gaussian separately
    for initial_guess in initial_guesses:
        # Fit the Gaussian to the data
        params, covariance = curve_fit(fgaussian, x_data, y_data, p0=initial_guess)

        # Append the parameters to the list
        params_list.append(params)

        # Generate the Gaussian fit using the obtained parameters
        y_fit = fgaussian(x_data, *params)

        # Append the fit to the list
        fits_list.append(y_fit)

        # Remove the fitted peak from the data for the next iteration
        y_data = y_data - y_fit

    return params_list, fits_list

# Example data
x_data_1 = first_col_1
y_data_1 = third_col_1

# Specify the parameters for each peak
# Format: [A, B, C, D] for each peak
parameters_to_fit = [
    [47130, 56, 25, 600],   # Parameters for Peak 1: [A, B, C, D]
    [11706, 154, 22, 700],   # Parameters for Peak 2: [A, B, C, D]
    [1284, 218, 1, 100],    # Parameters for Peak 3: [A, B, C, D]
    [1267, 519, 20, 200],    # Parameters for Peak 4: [A, B, C, D]
    [2832, 603, 20, 100],    # Parameters for Peak 5: [A, B, C, D]
]

# Fit multiple Gaussians to the data using the specified parameters
params_list, fits_list = fit_multiple_gaussians(x_data_1, y_data_1, parameters_to_fit)

# Calculate the error for each Gaussian fit
errors = []
for i, params in enumerate(params_list):
    y_fit = fits_list[i]
    residual = y_data_1 - y_fit
    mse = np.mean(residual**2)  # Mean Squared Error
    errors.append(mse)

# Plot the original data and the fitted Gaussians
plt.figure(figsize=(8, 6))
plt.scatter(x_data_1, y_data_1, label='Data', marker='.')
for i, params in enumerate(params_list):
    plt.plot(x_data_1, fits_list[i], label=f'Peak {i + 1}')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Print the parameters and errors for each Gaussian peak
for i, params in enumerate(params_list):
    A_fit, B_fit, C_fit, D_fit = params
    mse = errors[i]
    print(f'Peak {i + 1} Parameters:')
    print(f'A = {A_fit:.8f}')
    print(f'B = {B_fit:.8f}')
    print(f'C = {C_fit:.8f}')
    print(f'D = {D_fit:.8f}')
    print(f'Mean Squared Error = {mse:.8f}')
```
