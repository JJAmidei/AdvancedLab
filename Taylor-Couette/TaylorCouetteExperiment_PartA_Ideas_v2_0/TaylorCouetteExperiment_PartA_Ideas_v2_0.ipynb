## 1.0 Introduction and Guide

The Taylor-Couette experiment examines fluid flow between independently rotating cylinders. It is, in many respects, the "hydrogen atom" of fluid mechanics. By reducing the experimental  configuration to as fundamental a geometry and composition as possible, it is possible to make very precise measurements and do exacting tests of underlying theories. Astonishingly, despite the simplicity of the geometry, the system exhibits a very wide range of dynamical phenomena over the range of paramater variations that can be employed. Even when the outer cylinder is kept at rest, an increase in speed of the inner cylinder leads to a progression of flow states of increasing complexity in both space and time...ultimately producing highly turbulent flow that continues to be a focus of research in the field of fluid mechanics.

We will focus on the most fundamental flow state achieved in this system at low cylinder speeds and the observation that this state becomes unstable when critical values of the cylinder speeds are reached through careful control using computers. Flow visualization is a primary experimental technique that allows us to directly see and quantify features of the flow that develops when instability occurs.

The main ideas of this experiment can be summarized as follows:

1 The flow between two independently rotating co-axial cylinders has an exact solution to the equation of motion for moving fluids subject to "no-slip boundary conditions": the flow velocity much match the velocity of any walls with which it is in contact at the locations of the walls. Here the equation of motion, Newton's second law for fluids, is written as the *Navier-Stokes equation:* it expresses how the time variation of the flow velocity field is related to spatial derivatives of the velocity field and to forces imposed on the fluid, including viscous stresses due to internal friction within the moving fluid. In describing fluid motion, this equation must be accompanied by an equation of local conservation of mass (the "continuity equation") which for incompressible fluids is simply that the velocity field has zero divergence. You will see how the exact solition to these equations can be derived and how its coefficience are determined by the no-slip boundary conditions.

2 The exact solution described above can become unstable to small perturbations to the velocity field. This happens when, for a given rotational speed $\Omega_2$ of the outer cylinder, the rotational spped of the inner cylinder $\Omega_1$ reaches a critical value $\Omega_{1c}$. This critical value is typically expressed in terms of a critical value of the $Re_{c1}$ of the corresponding *Reynolds number* (Re) that is a dimensionless parameter defining the conditions imposed on the fluid. The definition of Reynolds number and its physical meaning will be explored in the discussion below. When instability occurs, a new flow state evolves with a pattern that has a spatial periodicity in the axial direction. This periodicity is typically expressed in terms of a "critical wavenumber" $k_c$ that appears in mathematical representations of the spatial periodicity of the perturbation to which the original flow state is most sensitive. We say that the original simple flow state has undergone a "pattern forming instability" and flow visualization techniques will be used to reveal and quantify the resulting patterns.

## 1.4 Forces acting within fluids

An important accomplishment of the Taylor-Couette experiment has been to test deductions made from how the forces that govern fluid motions are modeled.

Consider a  fluid element of volume $dxdydz$. The way in which forces are distributed throughout a moving fluid is central to understanding how fluids move. 

### 1.4.1 Gravitational force
One type of force, a body force like gravity, is due to some outside agency (e.g. earth's gravitational field) acting on any parcel of fluid. We can let the vector $\vec g$ represent the both the direction and the magnitude g of the force per unit mass due to gravity. Then the gravitational force acting on the fluid element with density $\rho$ is

$$ d\vec{F}^{(g)} = \vec{g} \rho dxdydz.$$

### 1.4.2 Pressure force

Next we should consider the force due to a pressure field p existing through the fluid. This pressure is isotropic and acts normal to any cross-sectional surface of the fluid. The net force on an element of fluid in the x-direction is given by:

$$  dF^{(p)}_x = \left(\frac{\partial p }{\partial x}dx\right) dydz $$

A similar expressions apply to the y and z components and thus in fector form, the pressure force on the fluid element is

$$ d\vec{F}^{(p)} = -\triangledown p dxdydz.$$

Another key class of forces is the internal frictional interaction between adjacent fluid elements as they move past one another. A key idea is that the frictional force on a surface element within a fluid layer undergoing shear is proportional to the spatial rate of change of the velocity component tangent to the surface element with respect to change in distance normal to the surface. This is Newton's model for viscous stresses.

### 1.4.3 Viscous force

A precise mathematical description of viscous stresses throughout a fluid requires the idea of tensors. The $i$'th cartesian component ($i$ = 1,2,3) of viscous stress (force per unit area) across a surface whose normal vector points in the $j$'th direction is denoted as $\sigma^{(v)}_{ij}$. Suppose an area element $dA$ has a vector normal to the area with magnitude equal to the area $dA$ and components $(dA_1,dA_2, dA_3)$. The fluid on the side of the area element opposite the normal vector will experience a force whose components are given by

$$dF_i = \sum_{j=1}^3\sigma^{(v)}_{ij}dA_j.$$

A key notion in describing viscous stresses is the idea of the rate-of strain tensor whose ij'th element is:

$$ \frac{1}{2}\left(\frac{\partial u_i}{\partial x_j} + \frac{\partial u_j}{\partial x_i}\right)$$

The off-diagonal elements $i \neq j$ describe how a fluid material element undergoes a pure shear with time.

A key empirical notion valid for many fluids (including air and water) is that the viscous stress is linearly related to this rate of strain tensor :

$$\sigma^{(v)}_{ij} = \hat\mu \left(\frac{\partial u_i}{\partial x_j} + \frac{\partial u_j}{\partial x_i}\right).$$

This equation is strictly valid only within the assumption that the fluid is incompressible so that $\triangledown \cdot u = 0$. For isotropic fluids like air and water the tensor must be symmetric under exhange of indices as shown.

Note here that a full account for the relationship between viscous stresses and the rate of strain within a fluid requires careful reasoning and considerations of symmetry. It is also useful to consider an underlying molecular interpretation of this viscous interaction. We refer you to the discussions in Guyon et al., Chapters 2 and 4 (see bibliography).

A net force on a fluid element will result if the surface forces vary across the fluid element. For a volume element $dxdydz$ the x-component of the viscous force is

$$dF_x^{(v)} = \left(\sigma^{(v)}_{11}(x+\frac{dx}{2})-\sigma^{(v)}_{11}(x-\frac{dx}{2})\right)dydz+\left(\sigma^{(v)}_{12}(y+\frac{dy}{2})-\sigma^{(v)}_{12}(y-\frac{dy}{2})\right)dxdz+\left(\sigma^{(v)}_{13}(z+\frac{dz}{2})-\sigma^{(v)}_{13}(z-\frac{dz}{2})\right)dxdy.$$

Using the definition of derivatives, this is equivalent to:

$$dF_x^{(v)} = \left(\frac{\partial\sigma^{(v)}_{11}}{\partial x}dx\right)dydz+\left(\frac{\partial\sigma^{(v)}_{12}}{\partial y}dy\right)dxdz+\left(\frac{\partial\sigma^{(v)}_{13}}{\partial z}dz\right)dxdy.$$

Substituting $\sigma^{(v)}_{11} = 2\hat\mu \frac{\partial u_x}{\partial x}$, $\sigma^{(v)}_{12} = \hat\mu\left( \frac{\partial u_x}{\partial y} + \frac{\partial u_y}{\partial x}\right)$, and $\sigma^{(v)}_{13} = \hat\mu\left(\frac{\partial u_x}{\partial z} + \frac{\partial u_z}{\partial x}\right)$, we get:

$$dF_x^{(v)} = \hat\mu\left[2\frac{\partial^2 u_x}{\partial x^2}+\left(\frac{\partial^2u_x}{\partial y^2}+\frac{\partial^2u_y}{\partial x \partial y}\right)+\left(\frac{\partial^2u_x}{\partial z^2}+\frac{\partial^2u_z}{\partial x \partial z}\right)\right]dxdydz.$$


$$dF_x^{(v)} = \hat\mu\left[\left(\frac{\partial^2 u_x}{\partial x^2}+\frac{\partial^2 u_x}{\partial y^2}+\frac{\partial^2 u_x}{\partial z^2}\right)+\frac{\partial}{\partial x}\left(\frac{\partial u_x}{\partial x}+\frac{\partial u_y}{\partial y}+\frac{\partial u_z}{\partial z}\right)\right]dxdydz$$

$$dF_x^{(v)} = \hat\mu\left[\triangledown^2 u_x + \frac{\partial}{\partial x}\triangledown \cdot \vec{u}\right]dxdydz.$$

But the assumption of incompressibility requires $\triangledown \cdot \vec{u} = 0$ so

$$dF_x^{(v)} = \hat\mu\triangledown^2 u_x \space dxdydz.$$

The same argument applies to the $y$ and $z$ components so that we can finally write:

$$d\vec{F}^{(v)} = \hat\mu\triangledown^2 \vec{u}\space dxdydz.$$




## 1.5 Governing equations

### 1.5.1 Continuity equation

The general equation for conservation of mass is called the continuity equation:

$$\frac{\partial \rho}{\partial t} = -\triangledown \cdot (\rho \vec {u}).$$

Under the assumption of incompressible flow $\rho$ = constant, this equation becomes

$$ \triangledown \cdot \vec{u} = 0.$$

### 1.5.2 Newton's second law and the Navier-Stokes equation

The physical acceleration of a parcel of fluid whose motion is described by a space and time dependent velocity field $\vec{v}(x,y,z,t)$ must account for the explicit time dependence of the field as well as the fact that the particle moves to a new location with a different velocity.

This acceleration is descripted by the *convective derivative*:

$$\frac{D \vec{u}}{dt} \equiv \frac{\partial \vec{u}}{\partial t}+(\vec{u} \cdot  \triangledown)\vec{u}.$$


Newton's 2nd Law for the acceleration of elements of fluid is then: 

$$\rho dxdydz \frac{D \vec{u}}{dt} = \Sigma (\text{ forces acting on an element of fluid}) = d\vec{F}^{(p)} + d\vec{F}^{(u)} + d\vec{F}^{(g)}.$$

When we insert actual expressions for the forces from above and divide through by the common factor $\rho dxdydz$ , we obtain the Navier Stokes equation. 

$$\frac{\partial \vec{u}}{\partial t}+(\vec{u} \cdot  \triangledown)\vec{u} =- \frac{1}{\rho}\triangledown p+\nu \triangledown^2 \vec{u}+\vec g.$$

Here $\nu \equiv \frac{\hat \mu}{\rho}$ is the kinematic viscosity of the fluid. It is important to remember at this pount that in this form of the Navier-Stokes equation, we have also assumed that the fluid is incompressible.

The Navier-Stokes equation is nonlinear and, as a consequence, describes an incredibly rich range of phenomena - from simple laminar flows to fully-developed turbulence.

### 1.5.3 Cylindrical polar coordinates

Since the Taylor-Couette experiment uses a cylindrical geometry, it is most useful to express the governing equations in cyclindrical polar coordinates. With reference to fluid dynamics textbooks, the equations are expressed as follows.

#### Continuity equation

$$\frac{1}{r}\frac{\partial}{\partial r}({ru_r})+\frac{1}{r}\frac{\partial}{\partial \phi}u_{\phi}+\frac{\partial}{\partial z} u_z = 0.$$

#### Navier-Stokes equations

$$\frac{Du_r}{Dt} - \frac{u_{\phi}^2}{r} = -\frac{\partial p}{\partial r} + g_r + \nu \left[\triangledown^2 u_r - \frac{u_r}{r^2} - \frac{2}{r^2}\frac{\partial u_{\phi}}{\partial \phi}\right].$$

$$\frac{Du_{\phi}}{Dt} + \frac{u_{\phi}u_r}{r} = -\frac{1}{r}\frac{\partial p}{\partial \phi} + g_{\phi} + \nu \left[\triangledown^2 u_{\phi} - \frac{u_{\phi}}{r^2} + \frac{2}{r^2}\frac{\partial u_r}{\partial \phi}\right].$$

$$\frac{Du_z}{Dt}  = -\frac{\partial p}{\partial z} + g_z + \nu \triangledown^2 u_r .$$


where

$$\frac{D}{Dt} \equiv \frac{\partial}{\partial t}+u_r\frac{\partial}{\partial r}+\frac{u_{\phi}}{r}\frac{\partial}{\partial \phi} + u_z\frac{\partial}{\partial z}$$

and

$$\triangledown^2 \equiv \frac{\partial^2}{\partial r^2}+\frac{1}{r}\frac{\partial}{\partial r} + \frac{1}{r^2}\frac{\partial^2}{\partial \phi^2} + \frac{\partial^2}{\partial z^2}.$$

#### Vorticity

An important quantity in fluid mechanics is the *vorticity* $\vec{\omega}$:

$$\vec{\omega} \equiv \triangledown \times \vec{u}.$$

In cylindrical polar coordinates:

$$\omega_r = \frac{1}{r}\frac{\partial u_z}{\partial \phi} - \frac{\partial u_{\phi}}{\partial z}.$$

$$\omega_{\phi} = \frac{\partial u_r}{\partial z} - \frac{\partial u_z}{\partial r}  .$$

$$\omega_z = \frac{1}{r}\frac{\partial (ru_{\phi})}{\partial r} - \frac{1}{r}\frac{\partial u_r}{\partial \phi} .$$



Note: the  vorticity vector $\vec{\omega}$  to be confused with a scalar angular frequency $\omega$ used below in the stability analysis.


#### Stress tensors

$$\sigma_{rr} = -p + 2\rho\nu\frac{\partial u_r}{\partial r}.$$
$$\sigma_{\phi \phi} = -p +2\rho\nu\left(\frac{1}{r}\frac{\partial u_{\phi}}{\partial \phi}+\frac{u_r}{r}\right).$$
$$ \sigma_{zz}=-p+2 \rho\nu \frac{\partial u_z}{\partial z}.$$

$$\sigma_{r\phi}=\sigma_{\phi r} = \rho\nu \left( \frac{1}{r}\frac{\partial u_r}{\partial \phi}+\frac{\partial u_{\phi}}{\partial r} - \frac{u_{\phi}}{r}\right).$$
$$\sigma_{rz}=\sigma_{z r} = \rho\nu \left(\frac{\partial u_r}{\partial z}+\frac{\partial u_z}{\partial r}\right).$$
$$\sigma_{\phi z}=\sigma_{z \phi} = \rho\nu \left(\frac{1}{r}\frac{\partial u_z}{\partial \phi}+\frac{\partial u_{\phi}}{\partial z}\right).$$



## 1.6 Base flow solution

### 1.6.1 Assume steady flow, azimuthal and axial symmetry, and axially directed gravity

If the system is azimuthally and axially symetric, then derivatives with respect to $t$, $\phi$, and $z$ vanish.

The continuity equation then becomes:

$$\frac{1}{r}\frac{\partial}{\partial r}({ru_r}) = 0.$$

This requires $ru_r = $ constant, which can only satisfy the boundary conditions of zero radial velocity at the walls if $u_r = 0$ everywhere.

We then find also by inspection of the equations that the z-component of the velocity also vanishes everywhere:

$$u_z = 0.$$

Using this fact and the vanishing of $u_r$ and $u_z$, the Navier-Stokes equations with gravity in the axial direction ($g_z = -g$) become:

$$-\frac{u_{\phi}^2}{r} = -\frac{1}{\rho}\frac{\partial p}{\partial r}.$$

$$ 0 = \nu \left[\left(\frac{\partial^2}{\partial r^2}+\frac{1}{r}\frac{\partial}{\partial r}\right)u_{\phi} - \frac{u_{\phi}}{r^2} \right].$$

$$0 = -\frac{1}{\rho}\frac{\partial p}{\partial z} - g.$$

### 1.6.2 Solving the resulting system for a symmetric flow state

We find that the azimuthal equation of the three components of the Navier-Stokes equation is satisfied by

$$u_\phi(r) = A r + B \frac{1}{r}.$$

In order to match the no-slip boundary conditions that $u_\phi(r) = \Omega_1 r_1$ at $r = r_1$ and $u_\phi(r) = \Omega_2 r_2$ at $r = r_2$, we require:

$$ A = \frac{r_2^2\Omega_2 - r_1^2\Omega_1}{r_2^2-r_1^2} = \frac{\Omega_2 - \eta^2\Omega_1}{1-\eta^2}.$$

$$ B = \frac{r_1^2 r_2^2(\Omega_1-\Omega_2)}{r_2^2-r_1^2}=\frac{\Omega_1-\Omega_2}{1-\eta^2}r_1^2.$$

In dimensionless form:

$$\frac{u_\phi(\frac{r}{d})}{\Omega_1 r_1} = \left[\tilde{A} \frac{r}{d} + \tilde{B}\left(\frac{r}{d}\right)^{-1}\right] .$$

$$\tilde{A} = \frac{\mu-\eta^2}{1-\eta^2}\frac{1-\eta}{\eta}.$$

$$\tilde{B}=\frac{1-\mu}{1-\eta^2}\frac{\eta}{1-\eta}.$$

Finally, for the pressure: if gravity is in the z-direction, this leads simply to an additive term to the pressure:

$$ p(r,z) = \tilde{p}(r) - \rho g (z-h) + p_{atm}.$$

Here $p_{atm}$ is the atmospheric pressure at the top of the cylinder (z=h). The term $\rho g (z-h)$ is called the hydrostatic head of fluid in a gravitational field, building up to a maximum added pressure $\rho g h$ at the bottom of the cylinder (z=0).

The remaining term for the pressure can be obtained from the radial component equation, which can be re-written as

$$\frac{\partial \tilde{p}}{\partial r} = \rho \frac{u_{\phi}^2}{r} .$$

This can be integrated using the above expression for $u_{\phi}(r) = Ar + B/r$ but it is useful to interpret in derivative form, since this is a clear statement that the radial derivative of the pressure must provide the centripetal acceleration of the fluid circulating around the inner cylinder.

## 1.7 Internal viscous stresses and torque required

One of the original uses of the Taylor-Couette apparatus was for measuring viscosity, usually by only rotating the outer cylinder and then measuring the torque transferred to the inner cylinder.

Here we will consider the general case of either or both cylinders rotating.

For a fluid with density $\rho$ and kinematic viscosity $\nu $, the dynamic viscosity is $ \rho \nu $. This is often denoted by the symbol $\mu$ or the symbol $\eta$ but we have been using these symbols for the radius ratio and speed ratio. Thus in the expressions for shear stress below, we will use the combination $ \rho \nu $.



$$\sigma_{r\phi}=\sigma_{\phi r} = \rho \nu \left( \frac{\partial u_{\phi}}{\partial r} - \frac{u_{\phi}}{r}\right) = -\rho \nu\frac{2B}{r^2} = -2 \rho \nu \frac{\Omega_1-\Omega_2}{1-\eta^2}\frac{r_1^2}{r^2}.$$.

Evaluated at $r = r_1$,

$$\sigma_{r\phi}=\sigma_{\phi r} = 2\rho\nu\frac{1}{1-\eta^2}(\Omega_2-\Omega_1).$$

The torque on an inner cylinder of height h and radius r_1 is then

$$ \text{Torque on inner cylinder} = \sigma_{\phi r} (2 \pi r_1 h) r_1 = 4 \pi h r_1^2 \rho\nu\frac{1}{1-\eta^2}(\Omega_2-\Omega_1).$$


In terms of aspect ratio $\Gamma = h/d$ and Reynolds numbers $Re_1$ and $Re_2$:

$$ \text{Torque on inner cylinder} = (\rho\nu^2 r_1)  \frac{4\pi\Gamma}{1-\eta^2}(\eta Re_2-Re_1).$$

Note that the expression $\rho\nu^2 r_1$ sets the scale for torque. An exercise below explores this quantitatively.

## 1.8 Equations in non-dimensional form and the meaning of the Reynolds number

### 1.8.1 Setting up a dimensionless form of the Navier-Stokes equation

If we use fluids with different densities and viscosities, will we see qualitatively different results...even though the Navier Stokes and continuity equations still apply?

A powerful way to describe the expected phenomena across broad classes of fluids - ranging from air to water and beyond in viscosity - is to convert the governing equations to dimensionless form. When doing this, we obtain insight into the meaning of the Reynolds number as a key parameter that describes the relative importance of viscous and inertial effects in the flow.

We establish scales for length, velocity, and time using features of the particular experiment - in this case the geometry and imposed rotation speeds of the cylinders in the Taylor-Couette experiment.

Length scale = the gap between the cylinders): $L_{\text{ref}} \equiv d.$ 

Velocity scale = the surface velocity of the rotating inner cylinder: $V_{\text{ref}} = \Omega_1 r_1.$

Time scale = length scale / velocity scale: $T_{\text{ref}} = \frac{d}{\Omega_1 r_1}.$ 


So now we replace our physical variables with the following substitutions:

$$ x = \tilde{x} d, y = \tilde{y} d, z = \tilde{z} d, r = \tilde{z} 4.$$

$$\triangledown = \frac{1}{d}\tilde{\triangledown}.$$

$$ \vec{u} = \tilde{\vec{u}} \Omega_1 r.$$

$$ t = \tilde{t} \frac{d}{\Omega_1 r_1}.$$

If we substitute these into the Navier-Stokes equations and collect terms, we find the following result (proving this will be an exercise):

$$\frac{\partial \tilde{\vec{u}}}{\partial \tilde{t}}+(\tilde{\vec{u}} \cdot  \tilde{\triangledown})\tilde{\vec{u}} =- \frac{1}{\rho(\Omega_1 r_1)^2}\tilde{\triangledown} p+\frac{\nu}{\Omega_1 r_1 d} \tilde{\triangledown}^2  \tilde{\vec{u}}+\frac{d}{(\Omega_1 r_1)^2}\vec g.\space\space\space\text{ Equation **}$$

In the first term on the right-hand side of the equation, we can define a reference pressure:

$$ P_{\text{ref}} \equiv \rho(\Omega_1 r_1)^2 $$

and then let

$$ p = \tilde{p} P_{\text{ref}} = \tilde{p} \rho(\Omega_1 r_1)^2 .$$

In the second term on the right-hand side of the equation, we see the combination previously defined as the Reynolds number (actually the reciprocal):

$$ \frac{\nu}{\Omega_1 r_1 d} = \frac{1}{Re_1}. $$

Finally, in the third term we can define a parameter called the *Froude number* $Fr$ 
( https://en.wikipedia.org/wiki/Froude_number )

$$ Fr \equiv \frac{\Omega_1 r_1}{\sqrt{g d}},$$

so that the third term becomes

$$\frac{1}{{Fr}^2}\hat{g}$$

where $\hat{g}$ is simply a unit vector pointing in the direction of gravity.

Finally, our equation becomes:

$$\frac{\partial \tilde{\vec{u}}}{\partial \tilde{t}}+(\tilde{\vec{u}} \cdot  \tilde{\triangledown})\tilde{\vec{u}} =-\tilde{\triangledown} \tilde{p}+\frac{1}{Re_1} \tilde{\triangledown}^2  \tilde{\vec{u}}+\frac{1}{{Fr}^2}\hat{g}.$$

The meaning of this is that if we used differrent fluids and different sizes of apparatus but match the Reynolds number and the Froude number, we will get the same flow! In practice, matching both can be difficult, as seen when you watch movies of ships made with scale models plowing through the water - they look fake! Fortunately, in the Taylor-Couette experiment the effect of gravity is "absorbed" into an offset in the pressure and can be ignored as long as the assumption of constant density remains valid.

This equivalence of experiments by matching dimensionless numbers, and in particular by matching the Reynolds number, is an extremely powerful result.

### 1.8.2 Physical interpretations of the Reynolds number

We can interpret the Reynolds number in a couple of ways.

#### Relative importance of viscous force to inertial effects

We observe that the momentum per unit volume of the flow has a scale $\rho \Omega_1 r_1$ and thus can think of a force scale given by the momentum flux:

$$f_{\text{inertial}} \equiv (\rho \Omega_1 r_1)(\Omega_1 r_1) = \rho (\Omega_1 r_1) ^2.$$

Now the viscous force will have a scale given by product of the dynamic viscosity $\hat{\mu} = \rho \nu$ and a characteristic gradient in the velocity:

$$f_{\text{viscous}} \equiv \hat{\mu}\frac{\Omega_1 r_1}{d} = \rho \nu \frac{\Omega_1 r_1}{d}.$$

THus the relative importance of the viscous force to inertial effects is given by

$$\frac{f_{\text{viscous}}}{f_{\text{inertial}}}= \frac{\rho \nu \frac{\Omega_1 r_1}{d}}{\rho (\Omega_1 r_1) ^2} = \frac{\nu}{\Omega_1 r_1 d} = \frac{1}{Re_1}.$$

The larger the Reynolds number, the less effective the viscosity is in dominating the flow. Conversely, for low Reynolds numbers ($Re_1 << 1$) viscosity is highly dominant, as is the case with flow around and inside microscopic organisms.

#### Relative time scales

Another interpretation considers time scales. An intertial time scale is

$$T_{\text{inertial}} = \frac{d}{\Omega_1 r_1}.$$

A viscous time scale is

$$T_{\text{viscous}} = \frac{d^2}{\nu}.$$

THen the ratio of the intertial time to the time over which viscous effects are "felt" is

$$\frac{T_{\text{inertial}}} {T_{\text{viscous}}}= \frac{\frac{d}{\Omega_1 r_1}}{\frac{d^2}{\nu}} = \frac{\nu}{\Omega_1 r_1 d} = \frac{1}{Re_1}.$$ 

So if the Reynolds number is large, inertial times are much shorter than viscous times.

## 1.9 Instability of the basic flow

Now we come to the real "punch line" of the Taylor-Couette experiment: the prediction and observation of a hydrodynamic instability. When this occurs, the presumed highly-symmetric "base flow" solution to the Navier-Stokes and continuity equations is found to become unstable to perturbations and is replaced by a new flow state in which symmetry is broken and a more complex spatial (and possibly temporal) pattern appears.

### 1.9.1 Rayleigh stability criterion for inviscid fluid

A basic constraint on the stability of flow moving in circular motion  was established by Lord Rayleigh.

If a parcel of fluid at radius r moves out to a new radius $r + \delta r$ but retains its original angular momentum, the flow is unstable if the pressure gradient cannot maintain the necessary centripetal acceleration.

Imagine that a perturbation causes a thin circulating ring of fluid of mass $m$ to move out from position to a larger radius. In the absence of viscous stresses the ring must retain its angular momentum. If the ring aquires a new azimuthal velocity $v'_{\phi}$, conservation of angular momentum requires:

$$(r+\delta r)m u'_{\phi} = r m u_{\phi}(r)$$

$$u'_{\phi} = \frac{r}{r+\ \delta r}u_{\phi}(r).$$

$$u'_{\phi} \approx \left(1-\frac{\delta r}{r}\right)u_{\phi}(r).$$

Recall that the local pressure gradient is given by:

$$\frac{\partial p}{\partial r} = \rho \frac{u_{\phi}^2}{r} .$$

For stability, the local pressure gradient at $r+\delta r$  must equal or exceed the value needed to provide the centripetal acceleration for the displaced element. Otherwise it will continue to displace outwards.

$$\rho\frac{u_{\phi}^2(r+\delta r)}{r+\delta r} > \rho\frac{{u'_{\phi}}^2}{r+\delta r}.$$

Taylor expand the left side and substitute from above for $u'_{\phi}$ on the right side:

$$\left[u_{\phi}(r)+ \frac{du_{\phi}(r)}{dr}\delta r\right]^2 > \left(1-\frac{\delta r}{r}\right)^2u_{\phi}^2(r).$$

Keep terms to first order in $\delta r$:

$$ u_{\phi}^2(r) + 2 u(r) \frac{du_{\phi}(r)}{dr}\delta r > u_{\phi}^2(r) - 2\frac{u_{\phi}^2(r)}{r}\delta r.$$

$$ \frac{du_{\phi}(r)}{dr} > - \frac{u_{\phi}(r)}{r}.$$

$$ u_{\phi}(r) + r \frac{du_{\phi}(r)}{dr} > 0.$$

The fluid is stable if:

$$ \frac{d[ru_{\phi}(r)]}{dr} > 0.$$

This is called the **Rayleigh stability criterion**.

If we insert

$$ u_{\phi}(r) = Ar + \frac{B}{r} $$

the above criterion leads to the requirement for stability to be

$$ 2Ar > 0.$$

Using

$$ A = \frac{\mu - \eta^2}{1-\eta^2}\Omega_1$$

we then must have (assuming $\Omega_1$ is positive)

$$ \mu > \eta^2.$$

which translates to the fluid flow being stable if

$$ \Omega_1 < \frac{1}{\eta^2}\Omega_2.$$

< TO BE ADDED: Insert graph of $\Omega_1 - \Omega_2$ plane showing stable region.>

In terms of inner and outer Reynolds number, the stability criterion is

$$ Re_1 < \frac{1}{\eta}Re_2.$$

In particular, the stability criterion implies that for $\Omega_2 = 0$ the flow becomes unstable for any value of $\Omega_1$. Instead, we observe that one must reach a critical value $\Omega_{1c}$ for instability to occur, due to the effects of viscosity. So we must refine our arguments and try to (a) understand why the instability is delayed, and (b) quantitatively predict the value for $\Omega_{1c}$ ... or more importantly, the dimensionless value for a critical Reynolds number $Re_{1c}$.

### 1.9.2 A heuristic estimate of a critical rotation speed for instability of a viscous fluid

The above inviscid criterion can be improved to account for viscous effects and arrive at new criterion that requires a nonzero critical inner cylinder angular velocity for a given outer cyclinder rotation, including zero. We will find that there must be enough outward force acting on a displaced fluid element to overcome losses induced by viscosity.

Specifically, we will assert that the imbalance between the centripetal force required to maintain a displaced fluid ring in rotation and the actual local pressure gradient is opposed by viscous drag. Under this condition a terminal radial velocity will be reached as the ring exands outward. If the time for the ring to move a distance comparable to its (small) radius is equal to or longer than the time within which viscosity causes the relative motion between the ring and surround fluid to damp out, i.e. to diffuse away, then the flow remains stable because the excess angular momentum is diffused away. This will become our new stability criterion.

Here viscosity is acting in two ways. It creates the drag force on the outward motion of the ring. It also diffuses the excess circular momentum out into the surround fluid, reducing the imbalance with the local pressure gradient.

Let's estimate the imbalance, assuming the ring has moved a distance $\delta r = \xi$ equal to its thin cross-sectional radius. 

The ring of fluid of density $\rho$ with radius $r$ and cross-sectional radius $\xi$ has mass

$$ m = \rho 2 \pi r \pi \xi^2.$$

From the inviscid arguments above, for an inital ring radius r, the imbalance is given by

$$\Delta F = \rho 2 \pi^2 r \xi^2 \left[\frac{u_{\phi}'^2}{r} - \frac{u_{\phi}^2(r+\xi)}{r}\right].$$

Using the idea that for a sudden displacement, the original angular momentum is retained allows us to insert the expression for $u'$ that assumed (at least for a short time) angular momentum conservation.

$$u'_{\phi} \approx \left(1-\frac{\xi}{r}\right)u_{\phi}(r).$$

Meanwhile 

$$u_{\phi}(r+\xi) \approx u_{\phi} + \frac{du_{\phi}(r)}{dr}\xi.$$

Inserting these into the expression for $\Delta F$ and simplifying gives to lowest order in $\xi$:

$$\Delta F = -\rho 4 \pi^2 r \xi^3 \left(\frac{u_{\phi}(r)}{r}+\frac{du_{\phi}(r)}{dr}\right).$$

The ring has surface area $(2\pi r)(2\pi \xi)$. If the ring drifts outward with radial speed $u_{\text{drift}}$ there is a velocity gradient of roughly $\frac{u_{\text{drift}}}{\xi}$. Thus the drag force is:

$$ F_{\text{drag}} \approx \rho \nu \frac{u_{\text{drift}}}{\xi} 4 \pi^2 r \xi.$$

We equate this in equlibrim to the force imbalance.

$$\Delta F = F_{\text{drag}}.$$

$$-\rho 4 \pi^2 r \xi^3 \left(\frac{u_{\phi}(r)}{r}+\frac{du_{\phi}(r)}{dr}\right) = \rho \nu \frac{u_{\text{drift}}}{\xi} 4 \pi^2 r \xi.$$

We thus estimate the drift speed to be:

$$ u_{\text{drift}} \approx -\frac{1}{\nu}\frac{u_{\phi}}{r}\left(\frac{u_{\phi}(r)}{r}+\frac{du_{\phi}(r)}{dr}\right)\xi^3.$$

The circular base flow field is stable against such perturbations if the drift time is longer than the momentum diffusion time.

$$ t_{\text{drift}} > t_{\text{diffusion}} $$

$$ \frac{\xi}{u_{\text{drift}}} > \frac{\xi^2}{\nu}.$$

This rearranges to a stability criterion:

$$ \frac{u_{\text{drift}}\space\xi}{\nu} < 1.$$

Substituting for $u_{\text{drift}}$,

$$-\frac{1}{\nu^2}\frac{u_{\phi}}{r}\left(\frac{u_{\phi}(r)}{r}+\frac{du_{\phi}(r)}{dr}\right)\xi^4 < 1.$$

Let's express the arbitray and small length $\xi$ as a fraction $\alpha$ of the gap $d$, where $\alpha << 1$. The stability criterion becomes

$$-\frac{1}{\nu^2}\frac{u_{\phi}}{r}\left(\frac{u_{\phi}(r)}{r}+\frac{du_{\phi}(r)}{dr}\right)d^4 < \left(\frac{1}{\alpha}\right)^4.$$

If we substitute

$$ u_{\phi}(r) = Ar + \frac{B}{r}$$

we get:

$$-\frac{2}{\nu^2}A\left(A+\frac{B}{r^2}\right)d^4 < \left(\frac{1}{\alpha}\right)^4.$$

Using

$$ A = \frac{\Omega_2 - \eta^2\Omega_1}{1-\eta^2}$$

and

$$ B =\frac{\Omega_1-\Omega_2}{1-\eta^2}r_1^2$$

and choosing a "worst case" scenario $r = r_1$ (giving the largest possible value for $B/r^2$), the stability criterion becomes

$$ \frac{2\Omega_1d^4}{\nu^2}\frac{\eta^2\Omega_1-\Omega_2}{1-\eta^2} <\left(\frac{1}{\alpha}\right)^4.$$

It turns out the expression on the left is identical to the Taylor number Ta defined in the section on dimensionless parameters:

$$ \text{Ta} \equiv \frac{2\Omega_1d^4}{\nu^2}\frac{\eta^2\Omega_1-\Omega_2}{1-\eta^2}.$$

So our stability criterion becomes

$$ \text{Ta} < \left(\frac{1}{\alpha}\right)^4 \equiv \text{Ta}_c.$$

In fact, the rigorous analysis outlined below gives a precise criterion that for $\eta = 0.655$ and $\mu = 0$, the critical value of the Taylor number


$$ \text{Ta}_c = 2367 $$

If we had chosen $\alpha = 1/7$ so that our ring cross-sectional radius $\xi = d/7$ we would have obtained a value 2401, so this all seems amazingly reasonable given that we have not tried to solve the Navier-Stokes equations directly!


### 1.9.3 A rigorous  approach to predicting critical instability parameters

A direct solution to the equations is in fact possible with the aid of both careful analysis and computational techniques. A rigorous approach called *linear stability theory* tests whether or not a given solution to the Navier-Stokes equation is stable to infintessimal perturbations.

Under the assumption of "infinitely long" cylinders, we have a base solution:

$$U_r(r) = 0.$$

$$U_\phi(r) = A r + B \frac{1}{r}.$$

$$U_z(r) = 0. $$

and the pressure must satisfy

$$\frac{\partial P}{\partial r}(r,z) = \rho \frac{U_{\phi}^2}{r} .$$

(The z-dependence allows for a hydrostatic pressure component to offset the body force of gravity but plays no role in the dynamics of the flow as long as the density is constant. Thus there must be no temperature variation.)

We use capital letters to denote the base solution and then perturb it with small variations:

$$\vec{u}(r,\phi,z) = \vec{U}(r) + \tilde{\vec{u}}(r,\phi,z)$$.

$$p(r) = P(r,z) + \tilde{p}(r,\phi,z).$$

This is inserted into the Navier-Stokes equation and only terms to linear order in the perturbations are kept.

Carrying out this analysis reveals that a solution will be of the form:


$$\tilde{\vec{u}}(r,\phi,z) = \text{Real}\left[\bar{\vec{u}}(r)e^{st - i(kz - m\phi)}\right].$$

We refer to $k$ as the axial wavenumber and $m$ as the azimuthal wavenumber. In order for the flow to be uniquely defined as the angle $\phi$ wraps around the cylinder, $m$ must be integer valued (including zero).

For given choices of the governing speed parameters (either Taylor number Ta and speed ratio $mu$ or inner and outer Reynolds numbers $\text{Re}_1$ and $\text{Re}_2$), and for selected values of axial and azimuthal wavenumbers $k$ and $m$, this is an eigenvalue problem: only a particular choice of the complex eigenvalue $s \equiv \sigma + i \omega$ will allow a solution to be found for the function $\bar{\vec{u}}(r)$ that satisfies the no-slip boundary condition $\bar{\vec{u}}(r) = 0$ at the two boundary points $r=r_1$ and $r=r_2$. This must be found numerically by solving a multi-dimensional set of ordinary differential equations in the components of $\bar{\vec{u}}(r)$ as a so-called "two-point boundary value problem".

If $\sigma < 0$ a perturbation will exponentially decay and the base flow is stable. If $\sigma = 0$ the perturbation neither grows nor decays, and we say the base flow is marginally stable. If $\sigma > 0$ the perturbation grows exponentially until it is large enought for nonlinear effects to become important. We say the base flow is unstable and find that a new flow solution must be found that replaces it. A new solution to the full nonlinear problem must be found, either by direct numerical simulation or from analytical perturbation theory.

The marginal Reynolds number $Re_1$ is defined to be the value, for a given choice of $Re_2$ and of $k$ and $m$, when the threshold $\sigma = 0$ is crossed. We can plot this marginal value versus $k$ to get what is called a *marginal stability curve*. We typically find that there is a particular value $\tilde{k}_c(m)$ at which the transitional Reynolds number $Re_1$ has a minimum value $\tilde{Re}_{1c}(Re_2,m)$. We call this the *critical* Reynolds number and $k_c$ is called the critical axial wavenumber for the given value of azimuthal wavenumber $m$. We can then try different values of this azimuthal wavenumber to find which gives the lowest minimum of all of the $\tilde{Re}_{1c}$. 

The critical axial wave number $k_c$ then gives the expected size of the vortex rolls. When m=0 and the rolls are time-inendent *Taylor vortices*, the vortex size is given by

$$\Delta z_\text{vortex} = \frac{1}{2}\frac{2\pi}{k_c} = \frac{\pi}{k_c}.$$

When expressed as a ratio to the size $d$ of the gap between cylinders,

$$\frac{\Delta z_\text{vortex}}{d} = \frac{\pi}{k_c d}.$$

Finally, therefore, we obtain for each experimentally set value of the outer cylinder speed and hence of the value of $Re_2$, there is a unique selected critical pair of wavenumbers $k_c$ and $m_c$ and a single minimum critical Reynolds number ${Re}_{1c}$. Note that this process also selects a critical value of corresponding imaginary part $\omega_c$ of the eigenvalue $s$...which is observed in experiment as the frequency of time-dependent flow. It is customary to report the value of $omega_c$ as a ratio with the inner cylinder angular velocity $\Omega_1$. It is an experimentally accessible parameter because the time dependence of the flow is easily measured using the changes in brightness of a laser spot reflected from flow visualization materia. Fourier analysis of the resulting signal reveals the frequencies in the flow's time dependence, which should consist of the critical frquency $\omega_c$ and its higher harmonics.

### 1.9.4 Interpolating for arbitrary radius ratio

The following code allows a user to input a particular radius ratio and then the critical parameters $Re_{1c}$, $k_c$, and $Ta_c$ are provided by interpolation between tabulated values read in with thre previous code. (The previous code must be run before using the following.) THis allows new experimental systems to be used with an accurate "look-up" of the expected critical points. All of this is for a stationary outer cylinder: $Re_c = 0$.

```python
# This code interpolates the computed critical values for a user-input value of radius ratio.
# It assumes previous code cell has been run so that the data arrays are in memory.

eta_request = float(input('What value of eta?'))
eta_round = round(eta_request,3)
#print(eta_round)
#print((.99-eta_round)*1000.)
row = int(round((.99-eta_round)*1000.,0))
#print(row)

#if eta_round == eta_request:
#    Re1c = Re1c_arr[row]
if eta_round < eta_request:
    Re1c = Re1c_arr[row]+(Re1c_arr[row-1]-Re1c_arr[row])*(eta_request-eta_round)/0.001
    Tac = Tac_arr[row]+(Tac_arr[row-1]-Tac_arr[row])*(eta_request-eta_round)/0.001
    kc = kc_arr[row]+(kc_arr[row-1]-kc_arr[row])*(eta_request-eta_round)/0.001    
else:
    Re1c = Re1c_arr[row]-(Re1c_arr[row]-Re1c_arr[row+1])*(eta_round-eta_request)/0.001
    Tac = Tac_arr[row]-(Tac_arr[row]-Tac_arr[row+1])*(eta_round-eta_request)/0.001
    kc = kc_arr[row]-(kc_arr[row]-kc_arr[row+1])*(eta_round-eta_request)/0.001

print('Re1c = {0:.4f}'.format(Re1c))
print('Tac = {0:.2f}'.format(Tac))
print('kc*d = {0:.5f}'.format(kc))
```

### 1.9.5 Counter-rotating cylinders and spiral vortex flow

When the outer cyclinder rotates opposite from the inner cylinder, $Re_2 < 0$, calculations find that when $Re_2$ reaches a special "co-dimension-2" value $Re_{2}^{(0->1)}$ where the instability changes to give rise to an azimuhtal wavenumber $m=1$. THe resulting flow is called *spiral vortex flow* and looks like a moving barber pole. This spiral vortex flow is retained for even more negative values for $Re_2$ until another "co-dimension-2" value $Re_{2}^{(1->2)}$ where the azimuthal wavenumber jumps from 1 to 2 and a steeper spiral vortex flow is obtained.

(The terminology "codimension-2 refers to the fact that two independent parameters  must be varied to explore a particular type of transition, here one in which more than one mode of new flow is in competition.)

Interesting dynamics is observed very close to the co-dimension-2 points because multiple modes are competing. Special mathematical theory has predicted a wide variety of interesting phenomena in these regions, including a new flow which looks like undulating Taylor vortices and other flows that include chaotic flow.

Finally, and importantly, careful investigation has uncovered a possibility that more than one minimum can exist in the marginal stability curve of $Re_1$ versus $k$. At very special values of the governing parameters $\eta$ and $\mu$ these minima have equal depth and there will be a competition of different *axial* wavenumbers. This prediction is now under active investigation by students in our laboratories.
