```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
import scipy.stats as stats

def is_float(string): 
    try:
        float(string)
        return True
    except ValueError:
        return False

data1 = np.genfromtxt('data\data4_26_24\data_good\p420v1200s3.csv', delimiter=',', dtype=str)
data2 = np.genfromtxt('data\data4_26_24\data_good\p415v1300s3.csv', delimiter=',', dtype=str)
data3 = np.genfromtxt('data\data4_26_24\data_good\p417v1400s3.csv', delimiter=',', dtype=str)

v_data_1 = [float(row[0]) if is_float(row[0]) else np.nan for row in data1]
i_data_1 = [float(row[1]) if is_float(row[1]) else np.nan for row in data1]
sigma_i_1 = np.asarray(i_data_1, dtype=np.float64)*0.001
v_data_2 = [float(row[0]) if is_float(row[0]) else np.nan for row in data2]
i_data_2 = [float(row[1]) if is_float(row[1]) else np.nan for row in data2]
sigma_i_2 = np.asarray(i_data_1, dtype=np.float64)*0.001
v_data_3 = [float(row[0]) if is_float(row[0]) else np.nan for row in data3]
i_data_3 = [float(row[1]) if is_float(row[1]) else np.nan for row in data3]
sigma_i_3 = np.asarray(i_data_1, dtype=np.float64)*0.001

# constants
k = 1.380649E-23
e = 1.602176634E-19

# linear fit function 
def func(x, a, b): 
    return a*x + b
```

## Important Note:

----From what it seems like from this write up (https://download.tek.com/document/LowCurtMsmntsAppNote.pdf), we can estimate the error in the current to be roughly $0.01%$ of the current measurement. This has been calculated from our data above.----

I just used the variance from the matrix given to use through pcov in curve_fit to calculate the error. I think sigma_a is essentially the error in our current. 

### 1200 V, 420 mTorr, 3 V steps

```python
# 1200 - initial plot, non-fit 

plt.figure()
plt.plot(v_data_1, i_data_1)
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('voltage')
plt.ylabel('total current')
plt.title('1200 V, 420 mTorr, 3 V steps')
plt.show()
```

```python
I_is_1 = -1.284397e-05 # Ion saturation current for data1

I_data_1 = [x - I_is_1 for x in i_data_1]

#I_data_1
I_data_1_filtered = I_data_1[60:100]
V_data_1 = v_data_1[60:100]
logI_data_1 = np.log(I_data_1_filtered)

popt, pcov = curve_fit(func, V_data_1, logI_data_1, p0=[1.0, 1.0])

V_data_1arr = np.asarray(V_data_1, dtype=np.float64)
logI_fit_1 = func(V_data_1arr, popt[0], popt[1])

a, b = popt
sigma_a, sigma_b = np.sqrt(np.diag(pcov))
fit = f'ln(I) = ({a:.3f} +/- {sigma_a:.3f})* V + ({b:.3f} +/- {sigma_b:.3f})'
print("Equation of the fitted line:", fit)

plt.figure()
plt.errorbar(V_data_1, logI_data_1, yerr=sigma_a*100, fmt='.', color='b', label='error scaled by 100')
plt.plot(V_data_1arr, logI_fit_1, color='r', label= 'Linear fit for ln(I)')
plt.xlabel('voltage')
plt.ylabel('total current')
plt.title('semi-log 1200 V, 420 mTorr, 3 V steps')
plt.legend()
plt.show()

print(f'Electron temperature: {(e/(k*a)):.0f} K +/- {(e/(k*a))*sigma_a:.0f} K')
```
