```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
import scipy.stats as stats

def is_float(string): 
    try:
        float(string)
        return True
    except ValueError:
        return False

data1 = np.genfromtxt('data\data4_26_24\data_good\p420v1200s3.csv', delimiter=',', dtype=str)
data2 = np.genfromtxt('data\data4_26_24\data_good\p415v1300s3.csv', delimiter=',', dtype=str)
data3 = np.genfromtxt('data\data4_26_24\data_good\p417v1400s3.csv', delimiter=',', dtype=str)

v_data_1 = [float(row[0]) if is_float(row[0]) else np.nan for row in data1]
i_data_1 = [float(row[1]) if is_float(row[1]) else np.nan for row in data1]
sigma_i_1 = np.asarray(i_data_1, dtype=np.float64)*0.001
v_data_2 = [float(row[0]) if is_float(row[0]) else np.nan for row in data2]
i_data_2 = [float(row[1]) if is_float(row[1]) else np.nan for row in data2]
sigma_i_2 = np.asarray(i_data_1, dtype=np.float64)*0.001
v_data_3 = [float(row[0]) if is_float(row[0]) else np.nan for row in data3]
i_data_3 = [float(row[1]) if is_float(row[1]) else np.nan for row in data3]
sigma_i_3 = np.asarray(i_data_1, dtype=np.float64)*0.001

# constants
k = 1.380649e-23 # Joules/Kelvin
e = 1.602176634e-19 # Coulombs

# linear fit function 
def func(x, a, b): 
    return a*x + b
```

## Important Note:

----From what it seems like from this write up (https://download.tek.com/document/LowCurtMsmntsAppNote.pdf), we can estimate the error in the current to be roughly $0.01%$ of the current measurement. This has been calculated from our data above.----

I just used the variance from the matrix given to use through pcov in curve_fit to calculate the error. I think sigma_a is essentially the error in our current. 

$$$$

$$ I_e(V_P) = I_{es}e^{\frac{-q_{e}(V_P - V_B)}{kT_e}} $$

$$$$

$$ln(I_{e}) = \frac{q_{e}V_B}{kT_e} - \frac{q_{e}V_P}{kT_e} + ln(V_{es}) = \frac{q_{e}V_B}{kT_e} + b$$

$$$$

$$I_{es} = \frac{1}{4}e n_e v_{e}A_{probe} = e n_e \biggr( \frac{kT_e}{2\pi m_e} \biggr)^{1/2}A_{probe} $$

$$$$

$$ n_{e} = e I_{es} \biggr( \frac{kT_e}{2\pi m_e} \biggr)^{1/2}A_{probe}  $$

$$$$

$$ f_c = 9\sqrt{n_{e, max}} $$

### 1200 V, 420 mTorr, 3 V steps

```python
# 1200 - initial plot, non-fit 
# from this we can find the floating potential V_f
# and the ion and electron saturation currents (I_is and I_es). 
# The floating potential is the voltage where 
# the total current is equal to zero. 
# The ion saturation current is the negative current given 
# by the first plateau. The electron saturation current
# is given by the second plateau after the exponential transition. 

V_f_1_max = v_data_1[62]
V_f_1_min = v_data_1[63]

plt.figure()
plt.plot(v_data_1, i_data_1)
plt.axhline(y=0, color='black', linestyle='--')
plt.axvline(x=V_f_1_max, color='red', linestyle='--')
plt.axvline(x=V_f_1_min, color='red', linestyle='--')
plt.xlabel('voltage')
plt.ylabel('total current')
plt.title('1200 V, 420 mTorr, 3 V steps')
plt.show()

print(f'V_f_max = {V_f_1_max} V')
print(f'V_f_min = {V_f_1_min} V')
print(f'V_f = {(V_f_1_max + V_f_1_min)/2} V +/- {abs(V_f_1_max - V_f_1_min)/2} V')
print()
print(f'I_is_min = {min(i_data_1)} A')
print(f'I_is_max = {i_data_1[10]} A')
print(f'I_is = {(i_data_1[10] + min(i_data_1))/2:0.11f} A +/- {abs(i_data_1[10] - min(i_data_1))/2:0.11f} A')
print()
print(f'I_es_min = {i_data_1[100]} A')
print(f'I_es_max = {max(i_data_1)} A')
print(f'I_es = {(max(i_data_1) + i_data_1[100])/2:0.9f} A +/- {abs(max(i_data_1) - i_data_1[100])/2:0.9f} A')
```

```python
I_is_1 = -1.284397e-05 # Ion saturation current for data1

I_data_1 = [x - I_is_1 for x in i_data_1]

I_data_1_filtered = I_data_1[60:100]
V_data_1 = v_data_1[60:100]
logI_data_1 = np.log(I_data_1_filtered)

popt, pcov = curve_fit(func, V_data_1, logI_data_1, p0=[1.0, 1.0])

V_data_1arr = np.asarray(V_data_1, dtype=np.float64)
logI_fit_1 = func(V_data_1arr, popt[0], popt[1])

a, b = popt
sigma_a, sigma_b = np.sqrt(np.diag(pcov))
fit = f'ln(I) = ({a:.3f} +/- {sigma_a:.3f})* V + ({b:.3f} +/- {sigma_b:.3f})'
print("Equation of the fitted line:", fit)

plt.figure()
plt.errorbar(V_data_1, logI_data_1, yerr=sigma_a*100, fmt='.', color='b', label='error scaled by 100')
plt.plot(V_data_1arr, logI_fit_1, color='r', label= f'{fit}')
plt.xlabel('voltage (Volts)')
plt.ylabel('ln(total current)')
plt.title('semi-log 1200 V, 420 mTorr, 3 V steps')
plt.legend()
plt.show()

T_e1 = e/(k*a)
sigma_T_e1 = abs(sigma_a*(e/(k*a**2)))

print(f'Electron temperature (K): {T_e1:.0f} K +/- {sigma_T_e1:.0f} K')
print(f'Electron temperature (eV): {T_e1/11606:.2f} eV +/- {sigma_T_e1/11606:.2f} eV')
```

```python
# plasma potential 
# this is the voltage where we hit the peak plateau
# we can find this by plotting the derivative of the current 
# and find the value of the voltage around the maximum and the 
# next zero value

didv = np.gradient(i_data_1, v_data_1)
didv_max = max(didv)
#mask = np.abs(didv - didv_max).argmin()
max_index = np.argmax(didv)
min_index = 100
V_P_max = v_data_1[max_index]
V_P_min = v_data_1[min_index]

plt.figure()
plt.plot(v_data_1, didv)
#plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('voltage')
plt.ylabel('total current')
plt.title('voltage vs. di/dv (1200 V, 420 mTorr, 3 V steps)')
plt.axvline(x=V_P_max, color='red', linestyle='--')
plt.axvline(x=V_P_min, color='red', linestyle='--')
#plt.legend()
plt.show()

print(f'V_P_max = {v_data_1[max_index]} V')
print(f'V_P_min = {v_data_1[min_index]} V')
print(f'V_P = {(v_data_1[max_index] + v_data_1[min_index])/2} V +/- {abs(v_data_1[max_index] - v_data_1[min_index])/2} V')
```

```python
# calculations for the floating potential. 
# This is when the total current is zero
# and V_B < V_P. 

I_is_min = min(i_data_1)
I_is_max = i_data_1[10]

I_es_min = i_data_1[100]
I_es_max = max(i_data_1)

I_is = (I_is_max + I_is_min)/2
I_es = (I_es_max + I_es_min)/2

sigma_Iis = abs(I_is_max - I_is_min)/2
sigma_Ies = abs(I_es_max - I_es_min)/2

V_P = (v_data_1[max_index] + v_data_1[min_index])/2
sigma_V_P = abs(v_data_1[max_index] - v_data_1[min_index])/2

sigma_V_f_calc = np.sqrt( (sigma_a*(1/a**2)*np.log(-I_is/I_es))**2 + (sigma_Iis/(a*I_is))**2 + (sigma_Ies/(a*I_es))**2 + sigma_V_P**2 )

V_f_calc = (1/a)*np.log(-I_is/I_es) + V_P
print(f'The calculated floating potential using the electron temperature: {V_f_calc:0.2f} K +/- {sigma_V_f_calc:0.2f} K')
```

```python
# abandoned for now, may come back later

# electron density calculation 
# needs more work


d = 3.0e-3 # mm (probe diameter estimate from Merlino)
A = 2*np.pi*(d/2)**2 # area of probe 

m_e = 9.11e-31 # kg (electron mass)

v_e = np.sqrt(8*k*T_e1/(np.pi*m_e)) # electron speed
sigma_v_e = np.sqrt((sigma_T_e1*np.sqrt(2*k/(np.pi*m_e*T_e1)))**2)

n_e = (4*I_es)/(e*A*v_e) # electron density

sigma_n_e = np.sqrt((sigma_Ies*4/(e*A*v_e)**2 + (sigma_v_e*(2*I_es/(e*A*v_e**2)))**2))

P = 420*(1/1000)*(133.3) # mTorr to Pascals
R = 8.3145 #mol^-1 K^-1

molecular_density = P/(R*T_e1) 

print(f'electron density: n_e = {round(n_e)} +/- {round(sigma_n_e)}')
print(f'Total molecualr density: n/V = {molecular_density}')
```
