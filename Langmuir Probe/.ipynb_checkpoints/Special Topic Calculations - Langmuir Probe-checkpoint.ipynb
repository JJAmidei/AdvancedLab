About 50 km - 1000 km above sea level, the ionosphere is a part of the Earth's upper atmosphere that is composed of a plasma. This is mostly due to incident solar radiation which dislodges electrons from neutral gas atoms and molecules. As in our experiment, this generated plasma may be described in terms of the electron density and the electron and ion temperatures. To measure these paramteres, satellites equipped with Langmuir Probes are used [1]. 

The code in the cell below models the measurements that a probe in the ionosphere might take.
